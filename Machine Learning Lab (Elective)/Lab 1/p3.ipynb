In this exercise, you will read an image of your choice using the PIL library, display it, and then
convert it to a numpy array. You will then work with only red plane of the array to :
1. Extract a 32x32 patch from anywhere of your choice
2. Extract the maximum pixel values along each row and column of the patch
3. Generate a new patch whose pixel values are the cosine of the original patch values
and display it.

```python
# This program is done by CS20B1044 Avinash R Changrani
from PIL import Image
import numpy as np 
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = Image.open(r"./img.jpg")
#img.show()
plt.imshow(img)
plt.show()
```

```python
arr = np.asarray(img)
#print(arr)
#print(arr.size)
print(arr.shape)
row,col,plane = arr.shape
```

```python
arr = arr[:,:,0]
# im = Image.fromarray(np.uint8(temp))
# im.show()
```

```python
# a) Extract a 32x32 patch from anywhere of your choice
array = arr[400:432, 669:701]
im = Image.fromarray(np.uint8(array))
plt.imshow(im)
plt.show()
im.size
```

```python
# b) Extract the maximum pixel values along each row and column of the patch
# print(array)
rows_max = np.max(array,axis = 1)
cols_max = np.max(array,axis = 0)
print(rows_max)
print(cols_max)
```

```python
# c) Generate a new patch whose pixel values are the cosine of the original patch values and display it.
cos_arr = np.cos(array)
im = Image.fromarray(np.uint8(cos_arr)) 
plt.imshow(im)
plt.show()
```
