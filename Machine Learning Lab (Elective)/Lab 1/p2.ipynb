a) Generate 300 3-dimensional vectors that come from a normal distribution with
mean vector as $[1 \ 2  \ 1]^t$ and 3x3 covariance matrix as [5 0.8 -0.3; 0.8 3 0.6; -0.3 0.6 4]

```python

mean = [1,2,1]

cov = [[5, 0.8, -0.3], [0.8, 3, 0.6], [-0.3, 0.6, 4]] 

pts = np.random.multivariate_normal(mean, cov, size=300)
```

(b) Make scatter plots of x1 vs x2, x1 vs x3, and x2 vs x3. Explain whatever relationships
you can gather from these plots.

```python
# x1 vs x2
plt.plot(pts[:, 0], pts[:, 1],'.')

plt.axis('equal')
plt.xlabel("x1")
plt.ylabel("x2")


plt.show()
```

```python
# x1 vs x3
plt.plot(pts[:, 0], pts[:, 2], '.')

plt.axis('equal')
plt.xlabel("x1")
plt.ylabel("x3")

plt.show()
```

```python
# x2 vs x3
plt.plot(pts[:, 1], pts[:, 2], '.')

plt.axis('equal')
plt.xlabel("x2")
plt.ylabel("x3")

plt.show()
```

(c) Calculate the mean vector and the covariance matrix using the 300 generated points.

```python
# Mean 
print("Mean is " + str(pts.mean(axis=0)))
```

```python
# Covariance 
print("Co-variance is \n" + str(np.cov(pts.T)))
```
