```python
# This Assignment is done by CS20B1044 Avinash R Changrani
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("./housing.csv")

df.head()
```

```python
df.isnull()
```

```python
# Finding Missing data
df.isnull().sum()
```

```python
# Removing Missing data and cleaning it
df = df.dropna()

df.isnull().sum()
```

```python
#len(pd.unique(df['ocean_proximity']))
pd.unique(df['ocean_proximity'])
```

```python
# give values to categorical variables
df['ocean_proximity'] = df['ocean_proximity'].map({'NEAR BAY':1, '<1H OCEAN':2, 'INLAND':3, 'NEAR OCEAN':4, 'ISLAND':5})

df.head()
```

```python
# Normalising the data
# df = (df - df.min())/(df.max() - df.min())

# df
```

```python
# split the dataset into training and testing data
X = df.drop('median_house_value',axis=1)
y = df['median_house_value']

from sklearn.model_selection import train_test_split

X_train, X_test,y_train, y_test = train_test_split(X,y ,random_state=42, test_size=0.3)
```

```python
# Linear Regression without regularisation
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(X_train, y_train)
y_pred = regr.predict(X_test)

print("Coefficients: \n", regr.coef_)
print("Intercept: \n", regr.intercept_)
mse = np.mean((y_test - y_pred)**2)
print("mse: \n", mse)
print("Regr Score: \n", regr.score(X_test,y_test))
```

```python
# Linear Regression with regularisation (Lasso Regression)
from sklearn.linear_model import Lasso
lasso_model = Lasso(alpha=0.1)
lasso_model.fit(X_train, y_train)

lasso_model.score(X_test,y_test), lasso_model.score(X_train,y_train), lasso_model.coef_
```

```python
# Ridge Regression
from sklearn.linear_model import Ridge
ridge_model = Ridge(alpha=0.1)
ridge_model.fit(X_train, y_train)

ridge_model.score(X_test,y_test), ridge_model.score(X_train,y_train), ridge_model.coef_
```
